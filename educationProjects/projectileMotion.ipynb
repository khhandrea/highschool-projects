각도에 따른 포물선 운동의 궤적 알아보기
======

## 모듈 임포트
#### 코드에 필요한 모듈들을 임포트한다.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## 궤적 그리기 함수
#### 주어진 초기속도, 각도에 해당하는 그래프를 그리는 함수

```python
def drawProjectileOrbit(v_0, theta, index, color):
    th = math.radians(theta)
    g = 10
    # x = np.arange(0, math.floor(2*v_0*v_0*math.sin(2*th)/g), 0.1)
    x = np.arange(0, 2.6, 0.1)
    y[index] = math.tan(th)*x - (g/(2*v_0*v_0*math.cos(th)*math.cos(th)))*x*x
    y_ = 0*x
    plt.plot(x,y[index], color, label=str(theta))
    plt.plot(x,y_, 'brown')
    plt.xlabel('distance')
    plt.ylabel('height')
    title = 'projectile orbit'
    plt.legend(loc='upper right')
    plt.title(title)
```

## 실행
#### 30, 45, 90도의 각도를 가진 포물선의 궤적을 비교한다.

```python
y = [0, 0, 0]
drawProjectileOrbit(5, 30, 0, 'r')
drawProjectileOrbit(5, 45, 1, 'g')
drawProjectileOrbit(5, 60, 2, 'b')
plt.show()
```

## 결과
#### 30도, 60도의 투사체의 수평 이동거리는 같고, 45도일 때가 가장 셋 중 가장 멀리 나간다.
