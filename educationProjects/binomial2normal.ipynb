이항분포의 표본과 정규분포 비교하기
======

## 모듈 임포트
##### 코드에 필요한 module들을 import한다

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import scipy.stats as stats
import math
```

## 이항 분포의 표본 추출 함수

##### 실행 횟수와 확률을 입력받아 B(n, p)의 임의추출 결과를 array로 반환한다.

```python
def binomial(n, p):
    dist = []
    for i in range(n):
        case = 0
        for j in range(n):
            if random.random() < p:
                case += 1
        dist.append(case)
    return dist
```

## 히스토그램 그리기
##### array를 이용한 히스토그램을 생성한다.

```python
def drawDist(data):
    plt.hist(np.array(data), bins = np.arange(100))
    plt.show()
```

## 정규분포 그리기
##### 비교를 위한 정규분포 그래프를 생성한다.

```python
def drawNorm(m, v):
    mu = m
    variance = v
    sigma = math.sqrt(variance)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma))
    plt.show()
```

## 실행
##### 임의추출한 이항분포 히스토그램과 정규분포 그래프를 그린다.

```python
drawDist(binomial(100, 0.5)) # B(100, 0.5)의 이항분포 그리기
drawNorm(50, 25) # N(50, 25)의 정규분포의 표본 그리기
```

## 결과
##### 큰 수의 이항분포의 확률변수 그래프의 개형은 정규분포의 생김새와 유사하다는 것을 알 수 있었다.
