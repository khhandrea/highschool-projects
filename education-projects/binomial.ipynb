## 이항 분포의 표본 알아보기

```python
import random
def binomial(num, pr):
    dist = []
    for i in range(num):
        if random.random() < pr:
            dist.append(True)
        else:
            dist.append(False)
    return dist
```

## 파이 그래프 그리기

```python
import matplotlib.pyplot as plt
def drawPie(data):
    plt.pie(data, 
            labels=['true', 'false'], 
            autopct='%1.2f%%',
            startangle=90,
            textprops={'fontsize': 14})
    plt.axis('equal')
    plt.title('X ~ B(100, 0.5)', fontsize=20)
    plt.show()
```

## 작동

```python
data = [0, 0]
for case in binomial(100, 0.5):
    if case == True:
        data[0] += 1
    else:
        data[1] += 1
drawPie(data)
```
